## Random Walks

```python
import random
```

```python
position = 0
walk = [position]
steps = 1000 # Giving the walk a thousands steps
for i in range(steps): # Step through the for loop 
    step = 1 if random.randint(0, 1) else -1 # Append to step either 0, 1, or -1 randomly
    position += step 
    walk.append(position) # Append it to positions list
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.plot(walk[: 100]) # Plot just the first 100 values from the walk
```

```python
plt.plot(walk[101:200]) # Plot the second 100 to 200 values 
```

## 1000 Coin Flips 

```python
import numpy as np
import math
```

```python
nsteps = 1000
draws = np.random.randint(0, 2, size = nsteps)
steps = np.where(draws > 0,1, -1)
```

```python
walk = steps.cumsum()
```

```python
walk
```

```python
walk.min()
```

```python
walk.max()
```

```python
walk.mean()
```

## Simulating Many Random Walks 

```python
nwalks = 5000
nsteps = 1000
draws = np.random.randint(0, 2, size=(nwalks, nsteps))
steps = np.where(draws > 0, 1, -1)
walks = steps.cumsum(1)
```

```python
walks
```

```python
walks.min()
```

```python
walks.max()
```

```python
walks.mean()
```
